<b>Task 2 : Image Contours</b>

```python
#read image
import cv2
import matplotlib.pyplot as plt
import numpy as np
screw = cv2.imread("../data/screw.png")
plt.imshow(screw)
```

Q: draw & plot the contour (inner) --> Child of circle which is inside the right shape

```python
gray = cv2.cvtColor(screw, cv2.COLOR_BGR2GRAY)
ret, bin_img = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY)
cont_list, hierarchy = cv2.findContours(bin_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
smallest_contour = min(cont_list, key=cv2.contourArea)
```

```python
out = cv2.drawContours(screw, [smallest_contour], -1, 255, 2)
plt.imshow(out)
```

```python
out = cv2.drawContours(screw, [smallest_contour], -1, 255, -1)
plt.imshow(out)
```
